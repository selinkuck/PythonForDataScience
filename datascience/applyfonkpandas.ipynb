```python
def salary_category(salary):
    if salary < 50000:
        return "low"
    elif 50000<= salary < 80000:
        return "medium"
    else:
        return "high"
```

```python
df["Salary"].apply(salary_category)
```

```python
df["Salary_Category"]= df["Salary"].apply(salary_category)
```

```python
df
```

```python
def performance(experince):
    if experince > 10:
        return 1
    else:
        return 0
```

```python
df["Experience"].apply(performance)
```

```python
df["Adjusted"]= df["Experience"].apply(performance) + df["Performance_Score"]
```

```python
df["Formatted_Name"] = df["Name"].apply(lambda x : x.replace("_"," "))
```
